 ## <span style="color: blue;">Univariate Analysis</span>

#### Univariate analysis examines a single column of data to summarize its characteristics.



##### Statistics: Mean, median, mode, range, and standard deviation.
##### Visuals: Histogram, bar chart, boxplot, or pie chart.
#### For example, analyzing test scores in one column to find averages or visualize their distribution.

 ### <span style="color: Red;">Which Graph Will be Plot For Univariate Analysis</span>

#### Which Graph will be plot for univariate analysis depends of the column is catagorical or numerical

 ### <span style="color: Red;">Catagorical - Frequency Distribution Table</span>

##### We Make two columns one is person name and second is his preference where he want to spend his vacations

#### Here we catch every category and find the frequency count that how much people want to go on the beach etc 

##### If the data is categorical, then a bar chart is indeed the best graph for visualizing a frequency distribution table. It clearly shows the frequency of each category, making comparisons easy.  

 ### <span style="color: Red;">Relative Frequency</span>

##### a pie chart is ideal for visualizing the relative frequency of categorical data. It shows proportions as slices, making it easy to compare each category's share of the whole.

 ### <span style="color: Red;">Cumulative Frequency</span>

##### A cumulative frequency graph (ogive) is the best choice for visualizing cumulative frequency. It shows the running total of frequencies, making it easy to observe trends and determine medians or percentiles.
