 ## <span style="color: blue;">Measure Of Dispersion</span>

 ### <span style="color: Red;">Why we need Measure Of Dispersion?</span>

##### Without dispersion, you only know the central value (mean or median), but you don’t know how the data is spread around it.
#### <span style="color: green;">Simple Analogy</span>
##### Imagine you have two classes, one with students scoring close to each other and the other with students scoring very differently:
##### Class 1 (close scores): 90, 92, 91, 93, 94
##### Class 2 (spread-out scores): 70, 90, 60, 100, 80
##### Both classes might have the same average score, but Class 2 has more spread or dispersion. Measures of dispersion help show that.

 ## <span style="color: blue;">Range</span>

 ### <span style="color: Red;">Formula of Range</span>

#####  Range = Maximum Value − Minimum Value

### <span style="color: green;">Example</span>

##### Imagine a dataset
##### [3,7,10,15,20]
##### Maximum value is 20
##### Minimum value is 3
##### Range = 20 - 3 = 17

 ### <span style="color: Red;">Why use Range</span>

##### Easy to Calculate: It's the quickest way to understand the spread of the data.
##### Gives a Basic Idea of Variability: Helps understand the gap between the smallest and largest values.

 ### <span style="color: Red;">Limitations of The Range</span>

##### 1. Sensitive to Outliers: If there’s an extremely high or low value, the range can be misleading
##### 2. Doesn't Show Data Distribution: It only tells the spread, not how the data is distributed between the smallest and largest values.
#### Data Distribution
##### For example
##### Dataset A:[1,50]
##### range = 50 - 1 = 49
##### Dataset B:[1,20,30,45,50]
##### range = 50 - 1 = 49
#### In both datasets, the range is 49, but the data is very different
##### In Dataset A, there are only two values: 1 and 50
##### In Dataset B, the values are spread out more evenly between 1 and 50.
#### The range doesn’t tell you this difference. It only focuses on the smallest and largest values, ignoring the rest of the data.

#### Step no 1
##### Calculate the Mean = 2 + 4 + 6 + 8 / 4 = 5
#### Step no 2
##### finding the square difference of the mean
##### (2 - 5)² = 9
##### (4 - 5)² = 1
##### (6 - 5)² = 1
##### (8 - 5)² = 9
##### square difference [9,1,1,9]
#### Step no 3
##### calulate the population variance
##### σ² = 9 + 1 + 1 + 9 / 4 = 20 / 4 = 5
##### calulating the sample variance
##### s² = 9 + 1 + 1 + 9 / 4-1 = 20 / 3 = 6.67
#### Results
##### population variance is 5

 ## <span style="color: blue;">Standard Deviation</span>

### <span style="color: Red;">How to calculate the standard deviation</span>

##### Step 1 Calculate the Mean of the Data
##### Step 2 Find the Square difference each data point and the mean
##### Step 3 Find the Variance (avarage of squared difference)
##### Step 4 Take square root of the variance

### <span style="color: Red;">Formula of Standard Deviation</span>


##### The upper formula is for Population standard deviation
##### The lower formula is for Sample Standard Deviation

#### xi = each data point
#### μ = mean of the population
#### x bar = sample mean
#### N = total number of data points in the Population data
#### n = Number of data points in the sample data data

### <span style="color: Red;">Why Standard Deviation is useful?</span>

#### Interpretation in Same Units: Since standard deviation is in the same units as the data, it's easier to understand. For example, if you're working with exam scores, the standard deviation will be in points, which is more meaningful than the squared units you get with variance.
#### It helps compare how much two datasets differ from their means. A dataset with a high standard deviation has more spread, while a dataset with low standard deviation has values closer to the mean.
