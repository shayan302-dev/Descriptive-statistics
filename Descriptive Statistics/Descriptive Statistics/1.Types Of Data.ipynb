## <span style="color: green;">Catagorical Data / Qualitative Data</span>

 ### <span style="color: blue;">Nominal Data</span>

##### Nominal data is a type of categorical data used to label or name variables without implying any numerical value or order. It is purely qualitative and represents groups or categories.

 ### <span style="color: red;">Key Characteristics of Nominal Data</span>

##### 1 Labels Only: It identifies or names categories (e.g., gender, colors, or countries). Gender: Male, Female, Non-binary.
##### 2 No Order: The categories have no inherent ranking or order for example (Colors: Red, Blue, Green. There’s no logical order where one color is "greater" or "lesser" than the others)
##### 3 Cannot Perform Arithmetic: Mathematical operations (e.g., addition or subtraction) are meaningless.
##### 4 Analysis: Frequency counts, mode, and chi-square tests are commonly used. for example (A group of 20 people were asked about their favorite fruit. Frequency Count: Simply counts how often each category (fruit) appears in the dataset.)

 ### <span style="color: blue;">Ordinal Data</span>

##### Ordinal data is a type of categorical data where the categories have a meaningful order or ranking, but the differences between the categories are not measurable or consistent. It represents data that can be ranked but doesn’t quantify the gap between ranks.

### <span style="color: red;">Key Characteristics of Ordinal Data</span>

##### Order Matters: The categories have a logical sequence (e.g., Low < Medium < High). for example Customer Satisfaction: Very Unsatisfied, Unsatisfied, Neutral, Satisfied, Very Satisfied.
##### Unequal Intervals: The difference between ranks is not standardized or measurable. for example Education Level: High School, Bachelor's, Master's, PhD.
##### No Arithmetic Operations: You can’t perform meaningful calculations like addition or subtraction. for example You cannot add or subtract these values (e.g., "High School + Bachelor's Degree" doesn't make sense).

##  <span style="color: green;">Numerical Data / Quantitative Data</span>

### <span style="color: blue;"> Discreate Data </span>

##### Discrete data refers to data that can take on only specific, distinct values and is typically countable. In the context of tabular data, discrete data is represented in a column where each entry is a whole number or category that cannot be subdivided.

### <span style="color: red;">Key Characteristics of Discreate Data </span>

##### Countable: Discrete data can only take certain values (e.g., 0, 1, 2, 3)
##### Non-continuous: There are no intermediate values between the possible data points
##### Categorical or Numerical: Discrete data can either be categorical (e.g., number of students in a class, different types of products) or numerical (e.g., number of cars sold, number of books)

### <span style="color: blue;"> Contineous Data </span>

##### Continuous data refers to data that can take on an infinite number of values within a given range and can include fractions or decimals. It is measurable rather than countable, meaning that the values can vary continuously, allowing for infinite precision depending on the measurement.

### <span style="color: red;">Key Characteristics of Contineous Data </span>

##### Can take any value within a specific range (including decimals).
##### Infinite possibilities: Between any two values, there are infinitely many other possible values
##### Measurable: Continuous data is typically obtained through measurements, not counting.
